```python
import os

def rename_images(main_folders):
  for folder in main_folders:
    for filename in os.listdir(folder):
      # Extract the file extension and last part of the name
      file_extension = os.path.splitext(filename)[1]
      new_name = os.path.splitext(filename.split("_")[-1])[0] + file_extension

      # Construct the full paths for old and new filenames
      old_path = os.path.join(folder, filename)
      new_path = os.path.join(folder, new_name)

      # Rename the image
      os.rename(old_path, new_path)

# Replace with your actual folder paths
main_folders = ["/content/DataSet1/Defected", "/content/DataSet1/OK"]

# Call the function to rename images
rename_images(main_folders)

print("Images renamed successfully!")
```

#### **1.3. ImageDataGenerator**

**ImageDataGenerator**,Data augmentation is used to increase the size of training set. Through Data augmentation we want to prevent overfitting ,this refers to randomly changing the images in ways that shouldn’t impact their interpretation, such as horizontal flipping, zooming, and rotating
* **Rescale:** One of the many magnification parameters adjusts the pixel values of our image.
* **Shear_range:** counterclockwise shear angle in degrees
* **Zoom_range:** zoom
* **Horizontal_flip:** flip image horizontally

```python
import torch
import lpips
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim

def prepare(img_path):
    img = image.load_img(img_path, target_size=(224,224))
    x = image.img_to_array(img)
    x = x/255
    return np.expand_dims(x, axis=0)

def cal_periodic_gaussian(img_name):
  test_image_path = "/content/DataSet2/noisy-test/"+ img_name
  test_image = cv2.imread(test_image_path)
  test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)

  original_image_path = "/content/DataSet2/noise-free-test/"+ img_name
  original_image = cv2.imread(original_image_path)
  original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

  # Preprocess the image (resize and normalize)
  test_image = cv2.resize(test_image, (224, 224))  # Resize to match the input shape of your model
  test_image = test_image / 255.0  # Normalize pixel values

  # Expand dimensions to create a batch of size 1
  test_image = np.expand_dims(test_image, axis=0)

  # Use the model to predict the denoised image
  denoised_image = denoiser_model.predict(test_image)

  # Convert the RGB images to PyTorch tensors
  denoised_image = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX).reshape(224,224, 3)

  ground = cv2.imread("/content/DataSet2/noise-free-test/"+ img_name , cv2.IMREAD_COLOR)
  ground_rgb = cv2.cvtColor(ground, cv2.COLOR_BGR2RGB)

  # Convert the RGB images to PyTorch tensors
  ground_truth_tensor = torch.from_numpy(ground_rgb).permute(2, 0, 1).float() / 255.0
  denoised_tensor = torch.from_numpy(denoised_image).permute(2, 0, 1).float() / 255.0

  # Calculate LPIPS
  lpips_loss = lpips.LPIPS(net='alex')
  lpips_score = lpips_loss(ground_truth_tensor, denoised_tensor)

  # Calculate PSNR
  mse = np.mean((ground_rgb - denoised_image) ** 2)
  psnr_score = 20 * np.log10(255 / np.sqrt(mse))
  # Calculate SSIM
  ssim_score = ssim(ground_rgb, denoised_image, multichannel=True)

  return (round(lpips_score.item()*100,3),round(psnr_score ,3) ,round(ssim_score * 100 ,3)  )

def cal_salt(img_name):
  noisyimage = cv2.imread('/content/DataSet2/noisy-test/'+img_name , cv2.IMREAD_COLOR)

  # Apply median filter several times
  filtered_image = cv2.medianBlur(noisyimage, 3)  # Kernel size of 3
  filtered_image = cv2.medianBlur(filtered_image,3)
  filtered_image = cv2.medianBlur(filtered_image,3)

  # Convert BGR to RGB for displaying with Matplotlib
  filtered_image_rgb = cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB)

  ground = cv2.imread('/content/DataSet2/noise-free-test/'+img_name, cv2.IMREAD_COLOR)
  ground_rgb = cv2.cvtColor(ground, cv2.COLOR_BGR2RGB)

  # Convert the RGB images to PyTorch tensors
  ground_truth_tensor = torch.from_numpy(ground_rgb).permute(2, 0, 1).float() / 255.0
  denoised_tensor = torch.from_numpy(filtered_image_rgb).permute(2, 0, 1).float() / 255.0

  # Calculate LPIPS
  lpips_loss = lpips.LPIPS(net='alex')
  lpips_score = lpips_loss(ground_truth_tensor, denoised_tensor)

  # Calculate PSNR
  mse = np.mean((ground_rgb - filtered_image_rgb) ** 2)
  psnr_score = 20 * np.log10(255 / np.sqrt(mse))
  ssim_score = ssim(ground_rgb, filtered_image_rgb, multichannel=True)
  return (round(lpips_score.item()*100,3),round(psnr_score ,3) ,round(ssim_score * 100 ,3)  )
```

#### **1.4. CNN Model**

A Convolutional Neural Network (ConvNet) is a Deep Learning algorithm which can take in an input image, assign importance (learnable weights and biases) to various aspects/objects in the image and be able to differentiate one from the other.

```python
model = Sequential()
model.add(Conv2D(32, (5, 5), input_shape=input_shape, activation='relu', name="conv2d_1"))
model.add(MaxPooling2D(pool_size=(3, 3), name="max_pooling2d_1"))
model.add(Conv2D(128, (3, 3), activation='relu', name="conv2d_2"))
model.add(MaxPooling2D(pool_size=(2, 2), name="max_pooling2d_2"))
model.add(Conv2D(256, (3, 3), activation='relu', name="conv2d_3"))
model.add(MaxPooling2D(pool_size=(2, 2), name="max_pooling2d_3"))
model.add(Flatten(name="flatten_1"))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
model.summary()
```

'fit()' parameters:
* **train:** training data,
* **validation_data:** validation set,
* **shuffle:** change of location of data in each epoch,
* **verbose:** to be able to see the outputs during the training (0-> does not show, 1-> does)
* **epoch:** determines how many times the dataset will be trained by traversing the model
* **callbacks:** An object that can perform actions at various stages of training

```python

from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_checkpoint = ModelCheckpoint('best_CNN_model.h5', monitor='loss', save_best_only=True)
early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)

history = model.fit(
    Train_generator,
    steps_per_epoch=32,
    epochs=15,
    validation_data=Val_generator,
    validation_steps=4,
    verbose=1,
    callbacks=[ReduceLROnPlateau(monitor='loss', factor=0.3, patience=3, min_lr=0.000001), early_stopping, model_checkpoint],
    shuffle=True
)
```

```python
def cal_sensitivity(conf_matrix):
    TP = conf_matrix[1, 1]
    FN = conf_matrix[1, 0]
    sensitivity = TP / (TP + FN)
    return sensitivity

def cal_specificity(conf_matrix):
    TN = conf_matrix[0, 0]
    FP = conf_matrix[0, 1]
    specificity = TN / (TN + FP)
    return specificity
```

```python
from keras.models import load_model
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt

# Load the saved model
model = load_model('/content/best_CNN_model.h5')

# Evaluate the model on the test set
scores = model.evaluate(Test_generator)
print(f"Test Accuracy: {scores[1]*100}")

# Generate predictions for the test set
y_pred = model.predict(Test_generator)
y_true = Test_generator.classes

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred.argmax(axis=1))
print("\n[INFO] Confusion Matrix:")
print(conf_matrix)

# Calculate metrics
print("\n[INFO] classification report:")
print(classification_report(y_true, y_pred.argmax(axis=1), target_names=Test_generator.class_indices))

print(f"Sensitivity: "+ str(cal_sensitivity(conf_matrix)))
print(f"Specificity: "+ str(cal_specificity(conf_matrix)))


# Plot the confusion matrix
plt.figure(figsize=(4, 4))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.get_cmap('Blues'))
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(Test_generator.class_indices))
plt.xticks(tick_marks, Test_generator.class_indices, rotation=45)
plt.yticks(tick_marks, Test_generator.class_indices)
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_true, y_pred[:, 1])
roc_auc = auc(fpr, tpr)

plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train_Acc', 'Train_Loss'], loc='upper left')
plt.show()

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FP Rate')
plt.ylabel('TP Rate')
plt.title('(ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

#### **1.5. VGG16 Model**


**VGG16** Architecture consists of 16 layers.

* **include_top :** Whether to include 3 layers fully connected to the top of the network
* **weight:** checkpoint from which model is initialized

```python
def create_Base_model_using_VGG16():
    model = VGG16(
        weights = "imagenet",
        include_top=False,
        input_shape = (224,224, 3) # goruntu boyutu
        )
    for layer in model.layers:
      layer.trainable = False
    return model
create_Base_model_using_VGG16().summary()
```

```python
def add_new_layers():
    model = create_Base_model_using_VGG16()
    x = model.output
    x = tf.keras.layers.Flatten()(x)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dense(256, activation="relu")(x)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dense(64, activation="relu")(x)
    predictions = tf.keras.layers.Dense(num_classes, activation="softmax")(x)

    final_model = tf.keras.models.Model(
        inputs = model.input,
        outputs = predictions)

    final_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return final_model

add_new_layers().summary()
```
